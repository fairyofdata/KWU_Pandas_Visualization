```python
import pandas as pd
import matplotlib.pyplot as plt

output_csv = 'output_2.csv'

result_df = pd.read_csv(output_csv)

house_var_counts = result_df.groupby('주변 역 개수').size().reset_index(name='주택 변수 개수')

result_df = result_df.merge(house_var_counts, on='주변 역 개수', how='left')

plt.bar(result_df['주변 역 개수'], result_df['단위면적가격'], width=0.8)
plt.xlabel('주변 역 개수')
plt.ylabel('단위면적가격')
plt.title('주변 역 개수별 집값 분포')

max_prices = result_df.groupby('주변 역 개수')['단위면적가격'].max()

for i, row in result_df.iterrows():
    if row['단위면적가격'] == max_prices[row['주변 역 개수']]:
        plt.text(row['주변 역 개수'], row['단위면적가격'], str(row['주택 변수 개수']), ha='center', va='bottom')

plt.show()
```
