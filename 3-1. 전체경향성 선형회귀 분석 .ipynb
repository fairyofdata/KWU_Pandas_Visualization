```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

df = pd.read_csv('2df0a798c0fe6ef5.csv')
df = df[df['Distance'] <= 4]
df = df[df['단위면적가'] <= 4000]

# 선형 회귀 계산
slope, intercept, r_value, p_value, std_err = linregress(df['단위면적가'], df['Distance'])

# 산점도 그래프 그리기
plt.scatter(df['단위면적가'], df['Distance'])

# 선형 회귀선 그리기
x = np.linspace(df['단위면적가'].min(), df['단위면적가'].max(), 100)
y = slope * x + intercept
plt.plot(x, y, color='red', label='Linear Regression')

plt.xlabel('Price')
plt.ylabel('Distance')
plt.title('Distribution of Price and Distance')

plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

df = pd.read_csv('2df0a798c0fe6ef5.csv')
df = df[df['Distance'] <= 4.0]
df = df[df['단위면적가'] <= 4000]

# 선형 회귀 계산
slope, intercept, r_value, p_value, std_err = linregress(df['단위면적가'], df['Distance'])

# R 제곱 계산
r_squared = r_value ** 2
print("R-squared:", r_squared)
```
