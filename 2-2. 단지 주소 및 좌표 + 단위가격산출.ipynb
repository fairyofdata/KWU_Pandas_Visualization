```python
df = pd.read_csv("/content/drive/MyDrive/23-1 빅데이터프로그래밍 기말과제/01. 로우데이터/아파트단지 데이터셋.csv.csv")
df2 = pd.read_csv("/content/drive/MyDrive/23-1 빅데이터프로그래밍 기말과제/02. 1차가공데이터/2-1. 아파트별 단위면적가격.csv")
```

```python
# field10과 field11을 하나의 컬럼으로 합치는 함수
df['field10'] = df['field10'].astype(str)
df['field11'] = df['field11'].astype(str)
def combine_fields(row):
    return row['field10'] + ' ' + row['field11']

# 합친 결과를 새로운 컬럼으로 추가
df['field10'] = df.apply(combine_fields, axis=1)
```

```python
# 2. 아파트단지 데이터셋과 단지별 평균단위가격 에서 필요한 컬럼만 가져오기, 컬럼명 적절하게 변경
df = df[['field3', 'field10','_X',	'_Y']]
df = df.rename(columns={'field10': '도로명'})
df = df.rename(columns={'field3': '단지명'})
```

```python
# 중복된 단지명이 있는지 확인하고, 서로 다른 이름으로 변경하고, 그래도 중복된 게 있다면, 제거한다.
duplicate_count = df.duplicated(subset=['단지명'], keep=False).sum()
print("중복된 행 수:", duplicate_count)
```

```python
duplicate_rows = df[df.duplicated(subset=['단지명'], keep=False)]
print("중복된 행의 내용:")
print(duplicate_rows)
```

```python
# 중복된 행을 찾기 위한 기준 컬럼
key_column = '단지명'
# 중복된 행의 '단지명'에 '도로명'을 병기하여 값을 수정
dup_mask = df.duplicated(subset=key_column, keep=False)  # 중복된 행을 마스킹
df.loc[dup_mask, '단지명'] = df.loc[dup_mask, '단지명'] + ' ' + df.loc[dup_mask, '도로명']
```

```python
duplicate_rows = df[df.duplicated(subset=['단지명'], keep=False)]
print("중복된 행의 내용:")
print(duplicate_rows)
```

```python
df.drop_duplicates(subset='단지명', inplace=True)
```

```python
duplicate_rows = df[df.duplicated(subset=['단지명'], keep=False)]
print("중복된 행의 내용:")
print(duplicate_rows)
# 이로써 중복된 아파트단지명이 모두 제거되었다. 
```

```python
from geopy.geocoders import Nominatim

def get_coordinates(address):
    geolocator = Nominatim(user_agent="stn_to_apt")  # user_agent는 원하는 이름으로 설정할 수 있습니다.
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None
```

```python
# 주소를 좌표로 변환하여 새로운 컬럼 추가
df2[['위도', '경도']] = df2['단지명_y'].apply(lambda x: pd.Series(get_coordinates(x)))
print(df)
```
