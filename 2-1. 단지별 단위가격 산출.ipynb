```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')
df = pd.read_csv("/content/drive/MyDrive/23-1 빅데이터프로그래밍 기말과제/01. 로우데이터/아파트(매매)__실거래가_20230528193525_fixed.csv", encoding = 'euc-kr')
df.head()
```

```python
df = df[['단지명', '전용면적(㎡)', '거래금액(만원)', '도로명']]
df.head()
```

```python
df['거래금액(만원)'] = df['거래금액(만원)'].astype(str).str.replace(',', '')
df['전용면적(㎡)'] = df['전용면적(㎡)'].astype(str).str.replace(',', '')
```

```python
# 모든 문자열이 숫자로 변환 가능한지 확인
print(df['거래금액(만원)'].str.isnumeric().all())
print(df['전용면적(㎡)'].str.isnumeric().all())

# 숫자로 변환 가능한 값만 선택하여 변환
df['거래금액(만원)'] = df['거래금액(만원)'].astype(str).str.replace(',', '').astype(float).astype(int)
df['전용면적(㎡)'] = df['전용면적(㎡)'].astype(str).astype(float).astype(int)
```

```python
df['단위면적가격'] = df['거래금액(만원)'] / df['전용면적(㎡)']
df['단위면적가격'] = round(df['단위면적가격']).astype(int)
df.head()
```

```python
df['평균가격'] = df.groupby('단지명')['단위면적가격'].transform('mean')
df.head()
```

```python
df = df.drop_duplicates(subset='단지명', keep='first')
df.head()
```

```python
df2 = df[['단지명', '단위면적가격', '도로명']]
df2
```
