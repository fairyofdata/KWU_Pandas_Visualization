```python
import pandas as pd
import csv
import folium    # 지오
```

```python
df = pd.read_csv('/content/drive/MyDrive/23-1 빅데이터프로그래밍 기말과제/01. 로우데이터/집.csv.csv')
df_station = pd.read_csv('/content/drive/MyDrive/23-1 빅데이터프로그래밍 기말과제/01. 로우데이터/역.csv.csv')
df_station.drop_duplicates('field3')
```

```python
delname=[]
for i in range(1,7):
    if i ==4:
        continue
    delname.append('field'+str(i))
delname.append('_GC_TYPE')
df.drop(delname,axis =1)
```

```python
map_CB = folium.Map(location = [37.559943,126.975314],zoom_start=30)
for i,store in df.iterrows():
    folium.Marker(location=[store['_Y'],store['_X']],popup=store['field4'],icon=folium.Icon(color='red',icon='star')).add_to(map_CB)
for i,store in df_station.iterrows():
    folium.Marker(location=[store['_Y'],store['_X']],popup=store['field3'],icon=folium.Icon(color='blue',icon='star')).add_to(map_CB)
map_CB.save('map_CB.html')
```

```python
import pandas as pd
import csv


files = ['전국지하철 자료1.csv','경춘.csv', '공항.csv', '신분당.csv']

merged_df = pd.DataFrame()

for file in files:
    df = pd.read_csv(file, encoding='UTF-8')
    merged_df = pd.concat([merged_df, df], ignore_index=True)
merged_df.dropna('_X')
output_file = '통합3.csv'
merged_df.to_csv(output_file, index=False)

print("CSV 파일 합치기가 완료되었습니다.")
```
